```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
companies = ["tsla", "aapl", "mcd", "hd"]
```

```python
tsla_bal = yf.Ticker('tsla')
tsla_bal_data = tsla_bal.balance_sheet.to_dict()
```

```python
tsla_bal_df = pd.DataFrame.from_dict(tsla_bal_data, orient='index')
```

```python
tsla_bal_df = tsla_bal_df.transpose()
```
